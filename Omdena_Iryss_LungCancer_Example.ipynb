```python
import pandas as pd
import statsmodels.api as sm

# Load your datasets
dfs = []
for year in range(2015, 2021):
    df = pd.read_csv(f'dataset_{year}.csv')
    df['YEAR'] = year  # Add the year column
    dfs.append(df)
df = pd.concat(dfs)

# Preprocess your data
df = df[df['YEAR'].between(2015, 2020)]

# Separate data into training and test set
train_df = df[df['YEAR'].between(2015, 2019)]
test_df = df[df['YEAR'] == 2020]

dependent_var_train = train_df['TOTCHG']
dependent_var_test = test_df['TOTCHG']

independent_vars_train = sm.add_constant(train_df.drop(columns=['TOTCHG', 'YEAR']))
independent_vars_test = sm.add_constant(test_df.drop(columns=['TOTCHG', 'YEAR']))

# Fitting the model
ols_model = sm.OLS(dependent_var_train, independent_vars_train)
result = ols_model.fit()

# Print model summary
print(result.summary())

# Predicting for 2020
predictions = result.predict(independent_vars_test)

# Now 'predictions' will contain the predicted total cost for 2020.
```

```python
# Instantiate the model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Fit the model
rf_model.fit(independent_vars_train, dependent_var_train)

# Make predictions
rf_predictions = rf_model.predict(independent_vars_test)

# Evaluate the model
rmse = mean_squared_error(dependent_var_test, rf_predictions, squared=False)
print(f'Root Mean Squared Error: {rmse}')
```

```python
def get_user_input():
    # Initialize all PAY1_, PAY2_, and RACE_ values to 0
    pay_values = ['PAY1_Cobra', 'PAY1_Medicare', 'PAY1_None', 'PAY2_Cobra', 'PAY2_Medicare', 'PAY2_None', 'PAY1_Secondary', 'PAY2_Secondary']
    race_values = ['RACE_White', 'RACE_Asian', 'RACE_Black', 'RACE_Hispanic', 'RACE_Native American', 'RACE_Other']
    data = {**{pay: 0 for pay in pay_values}, **{race: 0 for race in race_values}}

    data.update({
        'AGE': int(input('Enter your age: ')),
        'DRG': 34.90,
        'NPR': int(input('Enter number of procedures: ')),
        'NCHRONIC': int(input('Enter number of chronic diseases: ')),
        'HOSPID': 0,  # default value
        'DXn': 0  # default value
    })

    # Ask about insurance
    insurance = input('What is your primary insurance? (Cobra/Medicare/None/Other): ').lower()
    if insurance == 'cobra':
        data['PAY1_Cobra'] = 1
    elif insurance == 'medicare':
        data['PAY1_Medicare'] = 1
    elif insurance == 'none':
        data['PAY1_None'] = 1

    # Ask about race
    race = input('What is your race? (Asian/Black/Hispanic/Native American/Other): ')
    race_key = 'RACE_' + race
    if race_key in data:
        data[race_key] = 1

    # Convert annual income to ZIPINC_QRTL
    income = int(input('Enter your annual income: '))
    if income <= 25000:  # This is just an example, adjust these thresholds as necessary
        data['ZIPINC_QRTL'] = 1
    elif income <= 50000:
        data['ZIPINC_QRTL'] = 2
    elif income <= 75000:
        data['ZIPINC_QRTL'] = 3
    else:
        data['ZIPINC_QRTL'] = 4

    # Initialize all conditions to 0
    for column in condition_mapping.values():
        data[column] = 0

    # Ask user for their conditions
    conditions = input('Enter all of your current or past medical conditions, separated by a comma: ')
    conditions = conditions.lower().split(',')

    # Set the corresponding columns to 1 for each condition
    for user_condition in conditions:
        user_condition = user_condition.strip()  # Remove any leading/trailing whitespace
        for condition, column in condition_mapping.items():
            if condition in user_condition:
                data[column] = 1

    # Ask user about sex
    sex = input('What is your sex? (Female/Male): ').lower()
    if sex == 'female':
        data['SEX_Female'] = 1
        data['SEX_Male'] = 0  # Set SEX_Male to 0 when SEX_Female is 1
    elif sex == 'male':
        data['SEX_Female'] = 0  # Set SEX_Female to 0 when SEX_Male is 1
        data['SEX_Male'] = 1

    # Ask user about hospital admission/recent release
    tran = input('Were you recently admitted to the hospital or recently released? (Yes/No): ').lower()
    if tran == 'yes':
        data['TRAN_IN'] = 1
        data['TRAN_OUT'] = 0  # Set TRAN_OUT to 0 when TRAN_IN is 1
    else:
        data['TRAN_IN'] = 0  # Set TRAN_IN to 0 when TRAN_OUT is 1
        data['TRAN_OUT'] = 1

    return data

user_data = get_user_input()
user_df = pd.DataFrame([user_data])

prediction = rf_model.predict(user_df)
print(f"Predicted total charge: {prediction[0]}")
```
